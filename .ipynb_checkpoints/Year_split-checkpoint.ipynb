```python
#import os
import csv
import pandas as pd
import numpy as np
#import matplotlib as plt

#credits_file = 'tmdb_5000_credits.csv'
movies_file = 'tmdb_5000_movies.csv'

#credits_data = pd.read_csv(credits_file)
movies_data = pd.read_csv(movies_file)
```

```python
#credits_df = pd.DataFrame(credits_data)
#credits_df.head()
```

```python
movies_df = pd.DataFrame(movies_data)
movies_df.head(3)
```

```python
movies_df['release_date'] = movies_df['release_date'].dropna(how='all') 
movies_df['Year'] = movies_df['release_date'].str.split('-').str[0]
movies_df['Year'] = pd.to_numeric(movies_df['Year'], errors='coerce')
movies_df.head(3)
```

```python
decade_bins = [0,1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020]
decade_labels = ["1910's","1920's","1930's","1940's","1950's","1960's","1970's","1980's","1990's","2000's","2010's"]
movies_df['Decade'] = pd.cut(movies_df['Year'], decade_bins, labels=decade_labels)
movies_df.head(3)
```

```python
movies_df['Decade'].value_counts()
```
